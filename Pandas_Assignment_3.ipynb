```python
# Q.6 
# Window functions in pandas are a powerful way to perform calculations on a subset of rows that are related to the current row. They are similar to the window functions in SQL, which allow us to apply aggregate, ranking, or value functions over a partition or a sliding window of data.
#There are four types of window functions in pandas:

#1.Rolling window: This type of window function applies an aggregation over a fixed or variable size window of values. For example, we can use the rolling() method to calculate the rolling mean, sum, standard deviation, etc. of a Series or DataFrame.
#2.Weighted window: This type of window function applies a weighted aggregation over a non-rectangular window of values. For example, we can use the rolling() method with a scipy.signal window function as the win_type parameter to calculate the weighted mean, sum, standard deviation, etc. of a Series or DataFrame.
#3.Expanding window: This type of window function applies an aggregation over an expanding window of values. For example, we can use the expanding() method to calculate the cumulative sum, mean, maximum, minimum, etc. of a Series or DataFrame. The expanding window starts from the first row and grows until it reaches the last row.
#4.Exponentially weighted window: This type of window function applies an exponentially weighted aggregation over an expanding window of values. For example, we can use the ewm() method to calculate the exponentially weighted moving average, variance, standard deviation, etc. of a Series or DataFrame. The exponentially weighted window assigns more weight to recent observations and less weight to older observations.
```
